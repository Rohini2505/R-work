```r
data = read.csv("yob2016.txt", header = FALSE, sep=",")#, as.is = FALSE)
```

```r
colnames( data ) <- c('Names', 'Sex', 'Birth')
```

```r
# sort the data in descending order
sorted_data <- data[order(-data$Birth),]
```

```r
# select top 10 most popular names
first_10 <- head(sorted_data, n=10)
```

```r
first_10
```

```r
#import the colorspace library
library("colorspace")
```

## Plot a bar chart

```r
my_bar = barplot(first_10$Birth, names.arg = first_10$Names,
                 col= rainbow_hcl(2)[as.factor(first_10$Sex)], 
                 space = c(1.4), border="blue", las=2,
                 xlab="Names", ylab="Number of Names", main="10 Most Popular Names" , ylim=c(0,22000))

# Add the text 
text(my_bar, first_10$Birth+400, paste(first_10$Birth) ,cex=0.7) 
legend("topright", legend = unique(first_10$Sex), col= rainbow_hcl(2)[unique(first_10$Sex)],
      pch=20 , pt.cex = 2, cex = 0.8, horiz = FALSE, inset = c(0.05, 0.05))
```
